```python
import numpy as np

class KMeans:
    def __init__(self, k=3, max_iters=100):
        self.k = k
        self.max_iters = max_iters

    def fit(self, X):
        # Randomly initialize k centroids
        centroids = X[np.random.choice(len(X), size=self.k, replace=False), :]

        for i in range(self.max_iters):
            # Assign each data point to its closest centroid
            distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
            labels = np.argmin(distances, axis=0)

            # Update centroids to the mean of each cluster
            new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(self.k)])

            # Stop if centroids don't move much
            if np.allclose(centroids, new_centroids):
                break

            centroids = new_centroids

        self.centroids = centroids
        self.labels = labels

        return self
```

```python
from sklearn.datasets import make_blobs

# Generate some random data
X,y= make_blobs(n_samples=100, centers=3, n_features=3, random_state=42)

# Create an instance of the KMeans class and fit the data
kmeans = KMeans(k=3, max_iters=100)
kmeans.fit(X)

# Get the final centroids and cluster labels
centroids = kmeans.centroids
labels = kmeans.labels
```
