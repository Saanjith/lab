```python
import pandas as pd 
df=pd.read_csv(r'/Users/cra/Downloads/archive-3/mnist_train.csv')
```

```python
df.shape
```

```python
x=df.drop('label',axis=1).values
y=df['label'].values
```

```python
x=(x-x.mean())/x.std()
```

```python
import numpy as np
cov=np.cov(x,rowvar=False)
```

```python
val,vect=np.linalg.eigh(cov)
ind=np.argsort(-val)
val=val[ind]
vect=vect[:,ind]
n_compnonents=784
vect=vect[:,:n_compnonents]
vect.shape
```

```python
pc=np.dot(x,vect)
```

```python
pc.shape[1]
```

```python
val=val/np.sum(val)
csum=np.cumsum(val)
```

```python
sum(csum<=0.95)
val=val[:n_compnonents]

```

```python
import seaborn as sns 

da=pd.DataFrame({'Shape' : np.arange(pc.shape[1]),'var' :val})
sns.barplot(x='Shape',y='var',data=da,color='blue')

```
